```python
py.fillna({'module':'None'}, inplace = True)

def convert_tolist(x): 
    pattern = r'\[|\]' # convert '['a' , 'b']' to list
    s = re.sub(pattern, '', x)
    s = s.split(',')
    s = [i.strip().strip("''") for i in s]
    return [i.split('.')[0] for i in s] # if sklearn.preprocessing save sklearn

py['module'] = py['module'].apply(lambda x: convert_tolist(x))
```

```python
def convert_mdate(df, col):
    df[col] = pd.to_datetime(df[col])
    month = df[col].apply(lambda x: x.month)
    year = df[col].apply(lambda x: x.year)                       
    df[col] = pd.to_datetime(pd.DataFrame(list(zip(year, month)), columns = ['year', 'month']).astype(str).apply(' '.join, 1), format='%Y %m')

def create_t(df, time_df):
    [convert_mdate(time_df, i) for i in ['bq_content_first', 'ml_commit_first']] # call convert_mdate function
    time_df.set_index('repo', inplace = True)

    first_commit = df.reset_index().groupby('repo').agg({'datetime':min}) # get first commit time
    first_commit.columns = ['first_commit']
    last_commit = df.reset_index().groupby('repo').agg({'datetime':max}) # get last commit time
    last_commit.columns = ['last_commit']
    return pd.concat([time_df, first_commit, last_commit], axis = 1)
```

```python
time_df = create_t(d, time_df)
time_df.head()
```

```python
from tqdm.auto import tqdm
import pandas as pd
import math

if __name__ == "__main__":


#     df = pd.read_csv('py_data(commit_record).csv')
    col = list(df)
    df = df.astype({'commit_count': 'Int64'})
    df.groupby('repo').apply(lambda x: x.sort_values(['commit_year', 'commit_month'], ascending=[True, True]))
    df['datetime'] = pd.to_datetime(df[['commit_year', 'commit_month']].astype(str).apply(' '.join, 1), format='%Y %m')
    df['datetime'] = df['datetime'].dt.strftime('%Y-%m')
    grouped = df.groupby('repo')

    temp_group = []
    for name, group in tqdm(grouped):
        

        # add missing year and month
        period = pd.period_range(start=group['datetime'].min(), end=group['datetime'].max(), freq='M').astype('str')
        group = group.set_index('datetime').reindex(period).fillna({'repo': name})
        group = group.rename_axis('datetime').reset_index()
        group['datetime'] = pd.to_datetime(group['datetime'])
        group['commit_year'] = group.datetime.dt.year
        group['commit_month'] = group.datetime.dt.month
        
#         # if after first ml commit commit is NaN, set 1
        index = group[group['ml_commit'] == 1].index  # get first ml happen index
        have_lst = [col for col in group if col.startswith('have')]  # get package column name
        group.loc[index[0]:, 'ml_commit']= 1 # fill 1 if after ml_commit is NaN
        for i in index:
            add_lst = [col for col in have_lst if group.loc[i, col] == 1] # get used package
            group.loc[i:, add_lst]= 1 # set 1 once after used package

        
        temp_group.append(group)
        
            

    df = pd.concat(temp_group)
    df = df[col]
    df.to_csv('df_commit_v3.csv', index = False)
#     df.to_csv('df_time_v2.csv', index=False)
```
