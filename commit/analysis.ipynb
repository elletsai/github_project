### Fill month and ML commit

```python
from tqdm.auto import tqdm
import pandas as pd
import math

if __name__ == "__main__":


#     df = pd.read_csv('py_data(commit_record).csv')
    col = list(df)
    df = df.astype({'commit': 'Int64'})
    df.groupby('repo').apply(lambda x: x.sort_values(['commit_year', 'commit_month'], ascending=[True, True]))
    df['datetime'] = pd.to_datetime(df[['commit_year', 'commit_month']].astype(str).apply(' '.join, 1), format='%Y %m')
    df['datetime'] = df['datetime'].dt.strftime('%Y-%m')
    grouped = df.groupby('repo')

    temp_group = []
    for name, group in tqdm(grouped):
        
        add_list = ['ml_commit', 'new_sha1', 'ml_module']
        # add missing year and month
        period = pd.period_range(start=group['datetime'].min(), end=group['datetime'].max(), freq='M').astype('str')
        group = group.set_index('datetime').reindex(period).fillna({'repo': name})
        group = group.rename_axis('datetime').reset_index()
        group['datetime'] = pd.to_datetime(group['datetime'])
        group['commit_year'] = group.datetime.dt.year
        group['commit_month'] = group.datetime.dt.month
        
        
        # if after first ml commit commit is NaN, set 1
        for index in list(group.loc[~group['ml_commit'].isnull()].index):
    
            ml = list(group.loc[index,group.any(0)].keys())
            add_list += [col for col in ml if col.startswith('have')] 
            index +=1
            while index in group.index:
                if math.isnan(group.loc[index, 'ml_commit']):
                    group.loc[index, add_list] = 1
        
                index +=1
        temp_group.append(group)
        
            

    df = pd.concat(temp_group)
    df = df[col]
    df.to_csv('df_commit_v2.csv', index = False)
#     df.to_csv('df_time_v2.csv', index=False)
```

### Commit Analysis
- total commit number before  using ml / number of months per repo before using ml

- total commit number after using ml / number of months per repo after using ml

```python
import pandas as pd
from tqdm.auto import tqdm
```

```python
df = pd.read_csv('final_python.csv')
df.head()
```

```python
df.fillna(0, inplace = True)
```

```python
grouped = df.groupby('repo')
before = []
after = []
for name, group in tqdm(grouped):
    group.reset_index(inplace = True, drop = True)
    index = group.loc[group['ml_commit']!=0].index[0]
    before.append(group.iloc[:index]['commit_count'].sum() / len(group.iloc[:index]))
    after.append(group.iloc[index:]['commit_count'].sum() / len(group.iloc[index:]))
    
    
```

```python
before = pd.Series(before, dtype=object).fillna(0).tolist()
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Bar(x = df['repo'].unique(), y=before, name='before'))
fig.add_trace(go.Bar(x = df['repo'].unique(), y=after, name='after'))
```

```python
freq = pd.DataFrame(data = list(zip(df['repo'].unique(), before, after)),
            columns = ['repo', 'before_freq', 'after_freq'])
```

```python
freq
```

```python
more = freq[freq['after_freq'] > freq['before_freq']]
fig = go.Figure()

fig.add_trace(go.Bar(x = more['repo'].unique(), y=more.before_freq, name='before'))
fig.add_trace(go.Bar(x = more['repo'].unique(), y=more.after_freq, name='after'))
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

def fuzzy_merge(df_1, df_2, key1, key2, threshold=90, limit=2):
    """
    df_1 is the left table to join
    df_2 is the right table to join
    key1 is the key column of the left table
    key2 is the key column of the right table
    threshold is how close the matches should be to return a match, based on Levenshtein distance
    limit is the amount of matches that will get returned, these are sorted high to low
    """
    s = df_2[key2].tolist()

    m = df_1[key1].apply(lambda x: process.extract(x, s, limit=limit))    
    df_1['matches'] = m

    m2 = df_1['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= threshold]))
    df_1['matches'] = m2

    return df_1
```
